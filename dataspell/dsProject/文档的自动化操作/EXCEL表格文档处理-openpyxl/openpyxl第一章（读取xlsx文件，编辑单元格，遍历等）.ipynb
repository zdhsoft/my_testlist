## 使用模块 openpyxl
* openpyxl只支持对xlsx格式的文件进行操作
* 官方文档 https://pypi.org/project/openpyxl/

```python
%load_ext autoreload
%autoreload 2
import openpyxl
```

```python
#读取活动工作表
path = r'D:\design\jupyterNotebook\EXCEL数据处理-pandas\file\员工综合素质评价表.xlsx'
wb = openpyxl.load_workbook(path)
sheet = wb.active#  # 获取保存时的活动工作表
print(sheet)
sheet2=wb[wb.active.title]
print(sheet2)
sheet_data=wb.get_sheet_by_name('Sheet1')#通过表名获取数据
```

### 工作簿

```python
#插入一个工作表
sheet1=wb.create_sheet("Mysheet",0)#0代表插入第一个位置，不写也行
#更改工作表名称
sheet.title='顽皮的狗'
#得到一个工作表
s1=wb['顽皮的狗']
#复制一个工作表
copysheet_1=wb.copy_worksheet(sheet)#复制功能只能在当前工作簿中操作！
copysheet_1.title='慵懒的猫'
#修改工作表的背景色：RGB
sheet.sheet_properties.tabColor = "1072BA"
wb.save(path)
#插看所有工作表名称
print(wb.sheetnames)
#关闭工作簿。--不关闭数据还会在里面无法继续操作
wb.close()
```
