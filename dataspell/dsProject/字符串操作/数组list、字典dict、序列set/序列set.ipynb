# 序列set

## 序列
### 1.序列的构成形式

```python
set1 = {"1", "2"}
print(type(set1))

list=[1,2]
set2=set(list)#将数组转化为序列
print(type(set2))
```

### 2.增

```python
a=set("python")#拆分字符串,随机排列
print(a)

s=set()
s.add("dorlolo")
print(s)

s.add("dorlolo")#再次导入相同内容，内容不会重复
print(s)
```

### 3.删
#### (1)discard:移除指定元素，不存在会报错

```python
s1=set("python")
s1.discard("h")
print("s1:",s1)
```

#### (2)remove:移除指定元素，不存在也不会报错

```python
s1=set("python")
s1.remove("h")
print("s1:",s1)
```

#### (3)pop:随机删除某个元素，并返回该元组。 
    ps：psp（）中无需带参数

```python
# 集合pop随机移除某个元素并且获取那个参数,集合pop参数
s1=set("python")
s2 = s1.pop()
print("s1:",s1)
print("s2:",s2)
```

#### (4)clear：清除所有内容

```python
s1=set("python")
s1.clear()#清楚所有内容
print(s1)

s2={1,2,3}
s2.clear()
print("s2:",s2)
```

### 4.改
#### （1）交集

```python
s3 = {11, 22, 33}
s4 = {44, 33, 22}
s5=s3.intersection(s4)#存在交集
print(s5)

s3 = {11, 22, 33}
s4 = {44, 55, 66}
s5=s3.intersection(s4)#不存在交集
print(s5)
```

```python
# 交集的判断
if s3.isdisjoint(s4):#判断是否两个序列存在交集:返回True False
    print(s3.intersection(s4))
```

#### （2）并集

```python
s3 = {11, 22, 33}
s4 = {44, 55, 66}
s5=s3.union(s4)
print(s5)
```

#### （3）差集

```python
s3 = {11, 22, 33}
s4 = {44, 55, 33}
s5=s3.difference(s4)
print(s5)
```

#### （4）对称差集

```python
s3 = {11, 22, 33}
s4 = {44, 55, 33}
s5=s3.symmetric_difference(s4)
print(s5)
```

### update 批量更新，可导入多个数组

```python
s3 = {11, 22, 33}
li = [21, 4, 2, 312]
s3.update(li)
print(s3)
```
