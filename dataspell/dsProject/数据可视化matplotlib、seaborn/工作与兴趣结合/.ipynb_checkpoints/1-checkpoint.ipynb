##  单元格取值方法——获取每天半小时的平均进线量，并取平均值</br>
.values[0][0]

```python

import datetime
import numpy
initv=17#初始行位置，程序做了误差识别，只要误差在正负5行内就行
addv=21#递增
correct=0#行数修正

#index时间不对，误差修正
def error_cell_event(f,row,targetTime,column):
    global initv,addv,correct
#     print('get targetTime:',targetTime)
    for i in range(-5,5):#在行数row的正负5行进行查找
#         print("row:%s"%(row+i))
        getvalue=str(f.iloc[[row+i],[0]].values[0][0])
        #因为第一列有标题或空值，可能转化时间会报错，用try
        try:
            getTime=datetime.datetime.strptime(getvalue,"%H:%M:%S")
#             print('get Time:',getTime)
            if getTime<=targetTime:
#                 print(True)
                value=f.iloc[[row+i],[column]].values[0][0]
                correct=i
                return value,correct
        except:
            return 0,0

def result_check(v):
    try:
        a=int(v)
        if 0<a<1000:
            return True
    except:
        return False

def get_values(f,timelist,column):
    #timelist:把之前做的每半小时的时间做基准数组进行循环
    #column：匹配到对应行后，获取第column列的数据
    global correct,addv,initv
    valueList=[]
    valuearrary={}
    for num,indexTime in enumerate(timelist):
        
#         print('indexTime:',indexTime)
        for i in range(maxRows):
            row=initv+addv*i+correct+num
        #     print(row)
            if row<=maxRows:
                #每半小时录入的时间有大概10分钟误差，比如不都是9点30，可能是9点40，所以要判断误差是否在十分钟内，如果不对需要在附近几行进行查找 
                targetTime=datetime.datetime.strptime(str(indexTime),"%H:%M:%S")+datetime.timedelta(minutes=10)
#                 print('targetTime:',targetTime)
                try:
                    getTime=datetime.datetime.strptime(str(f.iloc[[row],[0]].values[0][0]),"%H:%M:%S")
#                     print('getTime:',getTime)
                    if getTime<=targetTime and getTime>=datetime.datetime.strptime(str(indexTime),"%H:%M:%S"):
                        value=f.iloc[[row],[column]].values[0][0]
                        if result_check(value):
                            valuelist.append(value)
        #                 print("row:%s----value:%s----time:%s----targetTime:%s"%(row,value,getTime,targetTime))
                    else:
                        value,correct1=error_cell_event(f,row,targetTime,column)
                        correct+=correct1
                        if result_check(value):
                            valuelist.append(value)
        #                 print("row:%s----value:%s----time:%s----targetTime:%s"%(row,value+correct,getTime,targetTime))
                except:#时间转化报错，说明找的单元格不对，
                    value,correct1=error_cell_event(f,row,targetTime,column)
                    correct+=correct1
                    if result_check(value):
                        valuelist.append(value)
        #             print("row:%s----value:%s----time:%s----targetTime:%s"%(row,value+correct,getTime,targetTime))
            else:
                break
        print(valubelist)
        value_everage=round(numpy.mean(valuelist))
#         print(value_everage)
        valuearrary[indexTime]=value_everage#求平均值
#         print('valuelist:',valuelist)
        valuelist.clear()
        correct=0
    return valuearrary
```

```python
everage_callIn=get_values(f,timelist,4)
print(everage_callIn)
everage_accept=get_values(f,timelist,8)
print(everage_accept)

print(type(everage_callIn))
print(type(everage_accept))
```

# 动态显示每日的数据

关键代码</br>
plt.icon()#开启交互模式</br>
plt.pause(1)#暂停1s，用sleep会卡死

```python
import datetime
import numpy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


class show_data:
    def __init__(self):
        self.initRow = 17  # 初始行位置，程序做了误差识别，只要误差在正负5行内就行
        self.addv = 21  # 递增,每日输入的间隔
        self.row_correct = 0  # 行数修正
        self.index = []  # 创建索引
        self.timer = 0  # 记录循环次数
        for i in range(9, 20):
            self.index.append(datetime.time(i, 0))
            self.index.append(datetime.time(i, 30))
        self.index = self.index[1:19]

        self.init_time = datetime.datetime.strptime(str(self.index[0]), "%H:%M:%S")
        self.traget_firstCell_v = self.init_time + datetime.timedelta(minutes=10)

        self.day = ''
    def readfile(self, file=r"res/每日值班记录 -2020.xlsx", sheet="Sheet1", index='Name'):
        #         f=pd.read_excel(r"res/每日值班记录 -2020.xlsx","Sheet1")
        #         f.set_index('Name')
        #         f.dropna(subset=['Name'],inplace=True)#删除Name列带有空值的行
        self.f = pd.read_excel(file, sheet)
        #         self.f.set_index(index)
        self.f.dropna(subset=[index], inplace=True)  # 删除索引列带有空值的行
        self.maxRow = self.f.iloc[:, 0].size  # 获取最大行数

    def verify_error_cell(self):
        rangelist=[]#生成循环数组
        for i in range(1,6):
            rangelist.append(0-i)
            rangelist.append(i)
        inputv=False
        while True:
            for i in rangelist:  # 在行数row的正负5行进行查找
                getvalue = str(self.f.iloc[[self.initRow + i], [0]].values[0][0])
                # 因为第一列有标题，可能转化时间会报错，用try
                try:
                    getTime = datetime.datetime.strptime(getvalue, "%H:%M:%S")
                    #             print('get Time:',getTime)
                    if self.init_time<=getTime <= self.traget_firstCell_v:
                        self.row_correct=i
                        self.timer-=1
                except:
                    if not inputv:
                        print("can't find init row! argument initRow's value may be false.")
                        inputv=input("init row num in excel is:")
                        self.initRow=inputv-2
                        break
                    else:
                        print('还是找不到初始行，请确认后再执行此程序！')
                        break



    def check_value(self,y1,y2):
        def check(valueList):
            list=[]
            for i in valueList:
                # print(list)
                try:
                    num=int(i)
                    if 0<=num<=1000:
                        list.append(i)
                    else:
                        list.append(numpy.nan)
                except:
                    list.append(numpy.nan)
            return list
        y1_l=check(y1)
        y2_l=check(y2)
        return y1_l,y2_l
    def get_nextValues(self, rows=[4, 8]):
        row = self.initRow + self.addv * self.timer + self.row_correct
        if row + self.initRow <= self.maxRow:
            firstCell_v = datetime.datetime.strptime(str(self.f.iloc[[row], [0]].values[0][0]), "%H:%M:%S")
            if self.init_time <= firstCell_v <= self.traget_firstCell_v:
                # 准备数据
                endrow = row + 18
                self.day = str(self.f.iloc[[row - 3], [0]].values[0][0]).split(" ")[0]

                get_y1 = list(self.f.iloc[row:endrow, rows[0]])
                get_y2 = list(self.f.iloc[row:endrow, rows[1]])

                self.y1,self.y2=self.check_value(get_y1,get_y2)
                # self.y1,self.y2=get_y1,get_y2
                return True
            else:
                print("数据首行识别错误")
                self.verify_error_cell()
        else:
            return False


    def show(self, ):
        print('run')
        self.x = range(0, 18)
        plt.ion()  # 开启交互模式
        # 设置中文显示
        plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体设置-黑体
        plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
        sns.set(font='SimHei')  # 解决Seaborn中文显示问题


        while True:
            result=self.get_nextValues()
            if not result:
                break
            # 1 创建画布
            plt.figure(figsize=(18, 6), dpi=80)
            # 标题
            dataTitle = "%s每小时电话量" % self.day
            print(dataTitle)
            print('y1:', self.y1)
            print('y2:', self.y2)
            plt.title(dataTitle)  # 大标题
            plt.xlabel('时间')  # x轴标题
            plt.ylabel('电话量')  # y轴标题
            # 2 画线
            plt.plot(self.x, self.y1, '', label="半小时进线量")  # x为x轴长度，y1为数据
            plt.plot(self.x, self.y2, '', label="半小时接线量")

            plt.legend(loc='upper right')
            plt.xticks(self.x, tuple(self.index))  # 设置x轴的刻度值

            plt.show()
            self.timer += 1
            plt.pause(1)

    def get_emtpy(self):
        v=self.f.iloc[14,[0]].values[0]
        print(v)
        print(type(v))

handel = show_data()
handel.readfile()
handel.show()
# handel.get_emtpy()
```
